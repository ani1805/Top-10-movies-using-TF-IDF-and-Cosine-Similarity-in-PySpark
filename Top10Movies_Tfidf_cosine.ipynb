```
plot_summaries=sc.textFile("/FileStore/tables/plot_summaries.txt")
plot_summaries.take(10)
```

```
#Storing the value of total number of movie summaries
No_of_documents=plot_summaries.count()
No_of_documents
```

```
#Separating each summary and the wikipedia ID using \t
summaries_1=plot_summaries.map(lambda x:x.split("\t"))
summaries_1.collect()
```

```
# summaries_no_space=summaries_1.filter(lambda x:len(x)>0)
# summaries_no_space.collect()
```

```
#Importing the required libraries for preprocessing the data
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
#Removing stop words and splitting the words in the summaries using regular expression characters
stop_words = set(stopwords.words('english'))
summaries_2 = summaries_1.flatMap(lambda x: [(x[0], word) for word in re.split(r"[\[\]# ./\"'-:;|?*“”’‘_]", x[1]) if len(word) > 1 and word.lower() not in stop_words])
summaries_2 = summaries_2.map(lambda x: ((x[0], x[1].lower()), 1))
summaries_2.collect()
```

```
#No of words after removing stop words
summaries_2.count()
```

```
#Reducing the (Key, value) pairs to get the number of occurences of a word in each document
summaries_word_count=summaries_2.reduceByKey(lambda x,y: x+y)
summaries_word_count.collect()
```

```
#Here, we reassign the RDD to get the (Key, Value) pairs and also writing their count
new_summaries = summaries_word_count.map(lambda x: (x[0][1], (x[0][0], x[1], 1)))
new_summaries.collect()
```

```
#This gives us the number of documents with a given word
summaries_idf = new_summaries.map(lambda x: (x[0], x[1][2]))
summaries_idf.collect()
```

```
#Reducing the (key, value) pairs to get the number of documents with a given word
reduced_idf_summaries = summaries_idf.reduceByKey(lambda x,y:x+y)
reduced_idf_summaries.collect()
```

```
#Applying the tf-idf formulae taught in class to calculate the idf term: log10(N/n)
import math
idf = reduced_idf_summaries.map(lambda x : (x[0], math.log10(No_of_documents/x[1])))
idf.collect()
```

```
summaries_tf = new_summaries.map(lambda x: (x[0], (x[1][0], x[1][1])))
summaries_tf.collect()
```

```
#Joining the idf values with the frequency term
tf_idf_summaries = summaries_tf.join(idf)
tf_idf_summaries.collect()
```

```
#Here, we calculate the weight of each word in a document by mutiplying the calculated IDF with the number of occurences of that word in the document
tf_idf_weighted = tf_idf_summaries.map(lambda x: (x[1][0][0], (x[0], x[1][1], x[1][0][1], x[1][0][1] * x[1][1])))
tf_idf_weighted.collect()
```

```
#Importing the movies_metadata.tsv file. This will help us to associate each frequency term with the corresponding movie title
movies_name = sc.textFile("/FileStore/tables/movie_metadata.tsv")
movies_name.collect()
```

```
new_movies = movies_name.map(lambda x: x.split("\t"))
new_movies.take(10)
```

```
#Separating out the movie-id and the movie title out of the imported text file. We only need these two to complete our question
movie_title = new_movies.map(lambda x: (x[0], x[2]))
movie_title.take(10)
```

```
#Joining the calculated weights of each term with the movie title.
summaries_movies = tf_idf_weighted.join(movie_title)
summaries_movies.take(10)
```
